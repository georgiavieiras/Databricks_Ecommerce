```python
%sql
create schema if not exists ecommerce.silver
```

```python
# Lendo DataFrame da tabela na camada silver

avaliacoes_df = spark.table("ecommerce.bronze.avaliacoes")
```

```python
#Salvando DataFrame na camada silver
avaliacoes_df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("ecommerce.silver.avaliacoes")
```
