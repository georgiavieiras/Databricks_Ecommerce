```python
%sql
create schema if not exists ecommerce.silver
```

```python
# Lendo DataFrame da tabela na camada silver

categorias_df = spark.table("ecommerce.bronze.categorias")
```

```python
#Salvando DataFrame na camada silver
categorias_df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("ecommerce.silver.categorias")
```
