```python
%sql
create schema if not exists ecommerce.silver
```

```python
# Lendo DataFrame da tabela na camada silver

itens_pedido_df = spark.table("ecommerce.bronze.itens_pedido")
```

```python
#Salvando DataFrame na camada silver
itens_pedido_df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("ecommerce.silver.itens_pedido")
```
