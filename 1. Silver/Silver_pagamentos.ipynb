```python
# Lendo DataFrame da tabela na camada silver

pagamentos_df = spark.table("ecommerce.bronze.pagamentos")
```

```python
#Salvando DataFrame na camada silver
pagamentos_df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("ecommerce.silver.pagamentos")
```
